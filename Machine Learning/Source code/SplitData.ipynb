# Buat fungsi *split* data

```python
# Fungsi split data
def split_data(SOURCE, TRAINING, TESTING, SPLIT_SIZE):

  filtered = []

  for files in os.listdir(SOURCE):
    path = os.path.join(SOURCE, files)
    if os.path.getsize(path):
      filtered.append(files)
    else:
      print('{} is zero length, so ignoring.'.format(files))

  # Split citra secara random
  random_files = random.sample(filtered, len(filtered))
  split = int(SPLIT_SIZE * len(filtered))
  training = random_files[:split]
  testing = random_files[split:]

  # Copy dan paste citra ke folder dataset
  for files in training:
    copyfile(os.path.join(SOURCE, files), os.path.join(TRAINING, files))

  for files in testing:
    copyfile(os.path.join(SOURCE, files), os.path.join(TESTING, files))
```
